```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup

posted_date=[]
scholarship_name=[]
deadline=[]

url="https://hec.ac.mw/scholarships"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html5lib')
#print(soup.prettify())

block = soup.find_all('div', class_="scholar-content-box")

for item in block:
    key = item.find('p', class_="scholar-meta").get_text(strip=True)
    posted_date.append(key)
    
for i in block:
    list_items=i.get_text().split('\n')
    y = (list_items[-4].strip()).split('.')
    scholarship_name.append(y[0].strip())
    deadline.append(y[1].strip())
    
for i in range(len(posted_date)):
    posted_date[i]=posted_date[i].lstrip('Posted on ')
    posted_date[i]=posted_date[i].rstrip(' by HEC Media')
    deadline[i] = deadline[i].lstrip('Closes 01 ')

df = pd.DataFrame({'Date Posted': posted_date, 'Scholarship Name': scholarship_name, 'Deadline': deadline})

df['Date Posted']=pd.to_datetime(df['Date Posted'])
df['Deadline']=pd.to_datetime(df['Deadline'])

print(df)
```
